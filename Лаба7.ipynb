```python
# Установка библиотек (если необходимо)
!pip install scikit-learn

# Импорт необходимых библиотек и модулей
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

Параметры согласно заданию
N - порядковый номер первой буквы фамилии (23 - Х)
M - порядковый номер первой буквы имени (15 - Н)

```python
# Фамилия начинается на "Х" (23-я буква), имя на "Н" (15-я буква)
N = 23
M = 15

print(f"Параметры:")
print(f"N (номер буквы фамилии) = {N}")
print(f"M (номер буквы имени) = {M}")
print()
```

```python
# 2. Загрузка набора данных load_wine
wine_dataset = load_wine()

# 3. Ознакомление с данными
print("Информация о наборе данных Wine:")
print(wine_dataset.DESCR[:500])
print()

# Вывод параметров вина
print("Параметры (признаки) вина:")
print(wine_dataset.data[:5])  # первые 5 строк
print()

# Вывод меток классов
print("Метки классов вина:")
print(wine_dataset.target[:20])  # первые 20 меток
print(f"Названия классов: {wine_dataset.target_names}")
print()

# 4. Разделение данных на обучающую и тестовую выборки
# Доля объектов в тестируемой выборке = N + 5
test_size_percent = N + 5
test_size = test_size_percent / 100

X_train, X_test, y_train, y_test = train_test_split(
    wine_dataset.data,
    wine_dataset.target,
    test_size=test_size,
    random_state=42
)

# Проверка размеров выборок
print("Разделение данных на выборки:")
print(f"Размер обучающей выборки: {len(X_train)}")
print(f"Размер тестовой выборки: {len(X_test)}")
print(f"Доля тестовой выборки: {test_size_percent} (N + 5 = {N} + 5)")
print()

# 5. Выбор и обучение моделей классификации
# Метод kNN с 7 ближайшими соседями
knn = KNeighborsClassifier(n_neighbors=7)
logreg = LogisticRegression(max_iter=1000)
ranfor = RandomForestClassifier()

# Обучение моделей
knn.fit(X_train, y_train)
logreg.fit(X_train, y_train)
ranfor.fit(X_train, y_train)

print("Модели обучены успешно")
print(f"Метод kNN использует {knn.n_neighbors} ближайших соседей")
print()

# 6. Оценка качества моделей
# Предсказания на тестовой выборке
y_pred_knn = knn.predict(X_test)
y_pred_logreg = logreg.predict(X_test)
y_pred_ranfor = ranfor.predict(X_test)

print("Фактические метки тестовой выборки:")
print(y_test)
print("Предсказания kNN:")
print(y_pred_knn)
print("Предсказания Logistic Regression:")
print(y_pred_logreg)
print("Предсказания Random Forest:")
print(y_pred_ranfor)

# Оценка точности моделей
print("Точность моделей (score):")
print(f"kNN: {knn.score(X_test, y_test):.4f}")
print(f"Logistic Regression: {logreg.score(X_test, y_test):.4f}")
print(f"Random Forest: {ranfor.score(X_test, y_test):.4f}")
print()

# 7. Классификация новых объектов
# Создание данных для новых объектов (бутылок вина)
# Для первой бутылки: первое число должно совпадать с N = 23
# Для второй бутылки: первое число должно совпадать с M = 15

# В наборе Wine 13 признаков, поэтому создаем объекты с 13 параметрами
# Остальные значения выбраны на основе анализа данных

# Создаем два новых объекта
X_new = [[23, 4, 3, 1, 2, 1, 3, 1, 1, 2, 2, 3, 2],
         [15, 3, 2, 2, 1, 2, 2, 2, 2, 1, 1, 2, 1]]

print("Классификация новых объектов:")
print(f"Бутылка 1 (первый параметр = N = {N}): {X_new[0]}")
print(f"Бутылка 2 (первый параметр = M = {M}): {X_new[1]}")
print()

# Предсказания для новых объектов
print("kNN:", wine_dataset.target_names[knn.predict(X_new)[0]], wine_dataset.target_names[knn.predict(X_new)[1]])
print("Logistic Regression:", wine_dataset.target_names[logreg.predict(X_new)[0]], wine_dataset.target_names[logreg.predict(X_new)[1]])
print("Random Forest:", wine_dataset.target_names[ranfor.predict(X_new)[0]], wine_dataset.target_names[ranfor.predict(X_new)[1]])

# knn.predict(X_new) - метод predict() предсказывает классы для новых объектов
# wine_dataset.target_names[...] - преобразуем числовой код класса в его название


# 8. Итог выполнения условий задания
print("\n" + "="*60)
print("ПРОВЕРКА ВЫПОЛНЕНИЯ УСЛОВИЙ ЗАДАНИЯ:")
print("="*60)
print(f"1. Количество ближайших соседей в kNN: {knn.n_neighbors} = 7 ✓")
print(f"2. Первая бутылка: первый параметр = {X_new[0][0]} = N = {N} ✓")
print(f"3. Вторая бутылка: первый параметр = {X_new[1][0]} = M = {M} ✓")
print("="*60)

# 9. Сохранение результатов (для Colab)
print("\nВсе условия задания выполнены успешно!")
```
