```csharp
public class Product
{
   private string name {get; set;} 
   private decimal price {get; set;}
   private int quantity {get; set;}
   protected string category {get; set;}
   internal string description {get; set;}
   
   public Product(string Name, decimal Price, int Quantity, string Category, string Description)
      {
         name = Name;
         price = Price;
         quantity = Quantity; // количество 
         category = Category;
         description = Description; // описание
      }

      public string Name
      {
         get { return name; }
         set { name = value;}
      }
      
      public decimal Price
      {
         get { return price; }
         set { price = value;}
      }

      public int Quantity
      {
         get { return quantity; }
         set { price = value;}
      }

      public decimal GetTotalPrice() // Общ. цена
      {
         return(price * quantity);
         Console.WriteLine($"Общая стоимость: {price * quantity}");
      }
      
      public void UpdateQuantity (int amount) // др. нов. количество
      {
         quantity += amount;
         Console.WriteLine($"Добавлен товар! ");
         Console.WriteLine($"Количество товаров обновлено! Количество:{quantity}");
      }

      public string GetProductInfo() // инфа о продукте
      {
         Console.WriteLine($"Описание товара:");
         return($"Название продукта: {name}, Цена: {price}, Количество: {quantity} шт., Категория товара: {category}, Описание: {description}");
      }
}

public class Seller
{
   private string name {get; set;}
   private string employeeID {get; set;}
   protected decimal salary {get; set;}
   internal string contactInfo {get; set;}

   public Seller(string Name, string EmployeeId, decimal Salary, string ContactInfo)
   {
      name = Name;
      employeeID = EmployeeId; // ID Сотрудника
      salary = Salary; // зп
      contactInfo = ContactInfo;
   }

   public string Name
   {
      get { return name; }
      set { name = value;}
   }

   public string EmployeeId
   {
      get { return employeeID; }
      set { employeeID = value;}
   }

   public decimal Salary
   {
      get { return salary; }
      set { salary = value;}
   }

   public string ContactInfo
   {
      get { return contactInfo; }
      set { contactInfo = value;}
   }
   
   List<Product> Productslist = new List<Product>();

   public void AddProduct (Product product) //  добавить продукт
   {
      Productslist.Add(product);
   }

   List<Seller> Sellerlist = new List<Seller>();

   public void SellProduct (Product product, int quantity) // умень кол прод на складе (проданный товар?)
   {
      if (Productslist.Contains(product))
      {
            product.UpdateQuantity(-quantity);
      }
   }

   public string GetSellerInfo() // Возварот инф о продавце?
   {
      Console.WriteLine($"Информация о продавце:");
      return($"Имя: {name}, ID: {employeeID}, Зарплата:{salary}, Контактная информация: {contactInfo}");
   }
}

public class Store
{
   private string storeName {get; set;}
   private string location {get; set;}
   public string storeHours {get; set;}

   public Store(string StoreName, string Location, string StoreHours)
   {
   storeName = StoreName;
   location = Location;
   storeHours = StoreHours;
   }

   public string StoreName
   {
      get { return storeName; }
      set { storeName = value;}
   }

   public string Location
   {
      get { return location; }
      set { location = value;}
   }

   public string StoreHours
   {
      get { return storeHours; }
      set { storeHours = value;}
   }

   List<Seller> Sellerslist = new List<Seller>();
   List<Product> Productslist = new List<Product>();

   public void AddSeller(Seller seller) // +продавец в магаз
   {
      Console.WriteLine($"Продавец добавлен в мазин:");
      Sellerslist.Add(seller);
   }

   public void ListProducts() // все товары в магазине
   {
      if (Productslist.Count == 0)
      {
         Console.WriteLine("В магазине нет товаров.");
      }
      else
      {
         foreach (Product product in Productslist)
         {
            Console.WriteLine(product.GetProductInfo());
         }
      }
   }

   public string GetStoreInfo() //инф. о магазе
   {
      Console.WriteLine($"Информация о магазине:");
      return($"Название магазина: {storeName}, Местоположение: {location}, График работы: {storeHours}");
   }
}

Product product = new Product("Плед", 2999, 2, "Интерьер", "Мягкое и пушистое покрывало синего цвета");
Console.WriteLine(product.GetProductInfo());
decimal totalPrice = product.GetTotalPrice();
Console.WriteLine($"Общая стоимость: {totalPrice} ");
Console.WriteLine();

product.UpdateQuantity(5);
Console.WriteLine();

Seller seller = new Seller("Константин", "1-0-1", 35000, "+7 (912) 345-67-89"); 
Console.WriteLine(seller.GetSellerInfo());
Console.WriteLine();
seller.AddProduct(product);

Store store = new Store("Ашан", "ул. Дмитрия Менделеева, 1", "с 8:00 до 23:00");
Console.WriteLine(store.GetStoreInfo());
store.ListProducts();
```
